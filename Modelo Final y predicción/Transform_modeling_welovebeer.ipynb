## Modelos sin transformación de datos

```
data_1 = data.copy()
data_1 = data_1.drop(['Año',	'Mes',	'SegmentoPrecio2',	'Marca2',	'Cupo2',	'CapacidadEnvase2',	'Volumen',	'disc',	'nr',	'Regional2',	'Gerencia2',	'SubCanal2',	'Categoria',	'Nevera'], axis=1)
data_1 = pd.get_dummies(data_1, columns=['Target'])
data_1 = data_1.rename(columns={"Target_1":"Marca1", "Target_2":"Marca2", "Target_3":"Marca3", "Target_4":"Marca_Inno1", "Target_5":"Marca_Inno2"})
data_1 = data_1.drop(["Target_0", "Unnamed: 0"], axis=1)
data_1
```

```
X = data_1.Cliente.values.reshape(-1,1)
y = data_1.drop(["Cliente"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Knn con 4 columnas target

```
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier(n_neighbors=100)
knn1.fit(X_train, y_train)

# Predecimos sobre nuestro set de entrenamieto
y_train_pred = knn1.predict(X_train)

# Predecimos sobre nuestro set de test
y_test_pred = knn1.predict(X_test)

# Comaparamos con las etiquetas reales
print('Accuracy sobre conjunto de Train:', accuracy_score(y_train_pred,y_train))
print('Curva Roc sobre el test set: ', metrics.roc_auc_score(y_test, y_test_pred))
```

```
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

knn1 = KNeighborsClassifier()
knn1.fit(X_train, y_train)

# Define los hyperparámetros

hp_candidates = [{'n_neighbors': np.arange(30, 60, 2),
                  'weights':['uniform','distance']}]

# Se busca mediante GridSearch el mejor hyperparámetro según los ya propuestos
grid = GridSearchCV(estimator=knn1, param_grid=hp_candidates, 
                    scoring="roc_auc", cv=5, verbose=5)

grid.fit(X_train, y_train)

# Se obtiene y se visualizan los mejor hiperparámetros encontrados con GridSearch
print(grid.best_score_)
print(grid.best_estimator_)
print(grid.best_params_)
```

```
knn1 = KNeighborsClassifier(n_neighbors = 29, weights = 'uniform')
knn1.fit(X_train, y_train)

# Predecimos sobre nuestro set de entrenamieto
y_train_pred = knn1.predict(X_train)

# Predecimos sobre nuestro set de test
y_test_pred = knn1.predict(X_test)

# Comaparamos con las etiquetas reales
print('Accuracy sobre conjunto de Train:', accuracy_score(y_train_pred,y_train))
print('Curva Roc sobre el test set: ', metrics.roc_auc_score(y_test, y_test_pred))
```
